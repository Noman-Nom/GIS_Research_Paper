#âœ… Step-by-Step: Random Forest Training

** âœ…Step 1: Load the Clean Dataset**

```python
import pandas as pd

# Load the cleaned dataset
file_path = '/content/drive/My Drive/qgis/cleaned_file.csv'
df = pd.read_csv(file_path)
print("âœ… Loaded data:", df.shape)
```

```python
import numpy as np

# Compute correlation matrix
corr_matrix = df.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Drop features with correlation > 0.98
to_drop = [column for column in upper.columns if any(upper[column] > 0.98)]
df_reduced = df.drop(columns=to_drop)

print(f"ðŸ“¦ Features dropped: {len(to_drop)}")
print(f"ðŸ” Remaining features: {df_reduced.shape[1]}")
```

```python
# Define features (X) and target (y)
X = df_reduced.drop(columns=['Label'])
y = df_reduced['Label']

# Train-test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)
print(f"âœ… Train size: {X_train.shape}, Test size: {X_test.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier

# Initialize and train model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

# Predict on test data
y_pred = rf_model.predict(X_test)

# Evaluation
print("ðŸ“Š Classification Report:")
print(classification_report(y_test, y_pred))

print("ðŸ§® Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
